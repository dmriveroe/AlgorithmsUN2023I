```python
from time import perf_counter
import math
import matplotlib.pyplot as plt
# Recursive Fibonacci execution time
def fibR(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fibR(n-1) + fibR(n-2)
```

```python
def time_recursive_fib(n):
    t0 = perf_counter()
    k = fibR(n)
    tf = perf_counter()
    return tf - t0
```

```python
ratio = (1 + math.sqrt(5))/2
timeList = []
timeList.append(time_recursive_fib(1))
timeList.append(time_recursive_fib(2))
t = time_recursive_fib(3)
timeList.append(time_recursive_fib(3))
for i in range(4, 51):
    timeList.append(t * ratio)
    t *= ratio

plt.plot(range(1, 51), timeList, label='Recursive Fibonacci')
```

```python
# Iterative Fibonacci (linear) execution time
def fibI(n):
    if n == 0:
        return 0
    a, b = 0, 1
    for i in range(1, n):
        a, b = b, a + b
    return b
```

```python
def time_iter_fib(n):
    t0 = perf_counter()
    k = fibI(n)
    tf = perf_counter()
    return tf - t0
```

```python
def fibA(n):
  r5= math.sqrt(5)
  return  (1/r5)*((1 + r5 )/2)**n - (1/r5)*((1 - r5 )/2)**n
```

```python
def time_closed_form_fib(n):
    t0 = perf_counter()
    k = fibA(n)
    tf = perf_counter()
    return tf - t0
```

```python
n_values = list(range(0, 51))
iterative_n_values = list(range(0, 1001))

closed_form_time = [time_closed_form_fib(n) for n in n_values]
iterative_time = [time_iterative_fib(n) for n in iterative_n_values]

plt.plot(n_values, closed_form_time, label='Closed-form Fibonacci')
plt.plot(iterative_n_values, iterative_time, label='Iterative Fibonacci')
plt.plot(iterative_n_values, iterative_n_values, label='y = n')

plt.xlabel('n')
plt.ylabel('Time (seconds)')
plt.legend()
```
