# Experimental analysis of the HeapSort execution time

```
def heapsort(arr):
    # Construir el montculo mximo
    n = len(arr)
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extraer elementos uno por uno
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Intercambiar el mximo con el ltimo elemento
        heapify(arr, i, 0)  # Llamar a heapify en el montculo reducido

def heapify(arr, n, i):
    largest = i  # Inicializar el nodo raz como el ms grande
    left = 2 * i + 1  # ndice del hijo izquierdo
    right = 2 * i + 2  # ndice del hijo derecho

    # Verificar si el hijo izquierdo es ms grande que la raz
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Verificar si el hijo derecho es ms grande que la raz y el hijo izquierdo
    if right < n and arr[right] > arr[largest]:
        largest = right

    # Si se ha encontrado un hijo ms grande, intercambiarlo y volver a llamar a heapify
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
```

```
A = [1,2,3,4,5,6,7,8]
heapsort(A)
print(A)

A = [6,7,2,5,4,3,8,1]
heapsort(A)
print(A)

A = [8,7,6,5,4,3,2,1]
heapsort(A)
print(A)
```


# Descripcin del Heap Sort y complejidad temporal

El algoritmo Heap Sort es un algoritmo de ordenacin basado en la estructura de datos conocida como heap (montculo). Consiste en construir un heap mximo a partir de la lista de elementos y luego extraer repetidamente el elemento mximo (raz del heap) y reconstruir el heap hasta que la lista est completamente ordenada.

El tiempo de ejecucin del Heap Sort se puede describir mediante las siguientes ecuaciones:

## Mejor caso - arreglo ordenado de forma creciente
En el mejor caso, cuando el arreglo ya est ordenado de forma creciente, el tiempo de ejecucin es:

$$t_b(n) = 5n\log_2{n} - 4 = \Theta(n\log{n})$$

## Peor caso - arreglo ordenado de forma decreciente
En el peor caso, cuando el arreglo est ordenado de forma decreciente, el tiempo de ejecucin es:

$$t_w(n) = \frac{3}{2} n\log_2{n} + \frac{7}{2}n - 4 = \Theta(n\log{n})$$

## Caso promedio - nmero esperado de comparaciones y intercambios en una permutacin aleatoria uniforme de $n$ elementos diferentes
En el caso promedio, considerando una permutacin aleatoria uniforme de $n$ elementos diferentes, el tiempo de ejecucin es:

$$t_a(n) = \frac{3}{4} n\log_2{n} + \frac{17}{4}n - 4 = \Theta(n\log{n})$$

Estas ecuaciones nos dan una idea de la complejidad temporal del algoritmo Heap Sort en diferentes casos. En todos los casos, el tiempo de ejecucin es de orden $\Theta(n\log{n})$, lo que hace que Heap Sort sea un algoritmo eficiente para ordenar arreglos grandes.

```
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(1, 21, 1)

plt.plot(n, 5*n*np.log2(n) - 4, label='Heap sort best time $5n\log_2{n} - 4$')
plt.plot(n, 3/4*n*np.log2(n) + 17/4*n - 4, label='Heap average time $3/4n\log_2{n} + 17/4n - 4$')
plt.plot(n, 3/2*n*np.log2(n) + 7/2*n - 4, label='Heap average time $3/2n\log_2{n} + 7/2n - 4$')
plt.xlabel('Array size')
plt.ylabel('Steps')
plt.legend(loc='upper left')

plt.show()
```

```
def heapify_steps(arr, n, i):
    steps = 0
    largest = i
    izquierda = 2 * i + 1
    derecha = 2 * i + 2

    # Verificar si el hijo izquierdo es mayor que el nodo actual
    if izquierda < n and arr[izquierda] > arr[largest]:
        largest = izquierda
        steps += 1

    # Verificar si el hijo derecho es mayor que el nodo actual
    if derecha < n and arr[derecha] > arr[largest]:
        largest = derecha
        steps += 1

    # Si el nodo actual no es el ms grande, intercambiar con el ms grande y continuar heapify
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        steps += 1
        steps += heapify_steps(arr, n, largest)
    return steps


def heap_sort_steps(arr):
    steps = 0
    n = len(arr)

    # Construir un max heap
    for i in range(n // 2 - 1, -1, -1):
        steps += heapify_steps(arr, n, i)

    # Extraer elementos del heap uno por uno
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        steps += heapify_steps(arr, i, 0)
        steps += 1
    return steps

```

Generate all permutations without repetitions of size *n* (for small *n*), plot the "exact" time  distribution (number of steps), and print the media and the standar deviation

```
from itertools import permutations
import matplotlib.pyplot as plt
import numpy as np
import math

n = 3
X = range(n)
times = []

for P in permutations(X):
  t = heap_sort_steps(list(P))
  times.append(t)
b = np.array(range(min(times)-1, max(times)+1)) + 0.5
plt.hist(times, bins=b)
plt.title("Heap Sort")
plt.xlabel("Steps")
plt.ylabel("Frequency")
plt.show()

print("n", n)
print("number of permutations n! =", math.factorial(n))
print("exp min =", min(times))
print("min formula $5n\log_2{n} - 4$ =", 5*n*np.log2(n) - 4)
print("exp max =", max(times))
print("max formula $3/2n\log_2{n} + 7/2n - 4$ =", 3/2*n*np.log2(n) + 7/2*n - 4)
print("exp mean =", np.mean(times))
print("mean formula $3/4n\log_2{n} + 17/4n - 4$ =", 3/4*n*np.log2(n) + 17/4*n - 4)
print("stdev", np.std(times))

```

# Sample of runs (montecarlo)

# Descripcin del Heap Sort y complejidad temporal

El algoritmo Heap Sort es un algoritmo de ordenacin basado en la estructura de datos conocida como heap (montculo). Consiste en construir un heap mximo a partir de la lista de elementos y luego extraer repetidamente el elemento mximo (raz del heap) y reconstruir el heap hasta que la lista est completamente ordenada.

El tiempo de ejecucin del Heap Sort se puede describir mediante las siguientes ecuaciones:

## Mejor caso - arreglo ordenado de forma creciente
En el mejor caso, cuando el arreglo ya est ordenado de forma creciente, el tiempo de ejecucin es:

$$t_b(n) = 5n\log_2{n} - 4 $$

## Peor caso - arreglo ordenado de forma decreciente
En el peor caso, cuando el arreglo est ordenado de forma decreciente, el tiempo de ejecucin es:

$$t_w(n) = \frac{3}{2} n\log_2{n} + \frac{7}{2}n - 4$$

## Caso promedio - nmero esperado de comparaciones y intercambios en una permutacin aleatoria uniforme de $n$ elementos diferentes
En el caso promedio, considerando una permutacin aleatoria uniforme de $n$ elementos diferentes, el tiempo de ejecucin es:

$$t_a(n) = \frac{3}{4} n\log_2{n} + \frac{17}{4}n - 4 = \Theta(n\log{n})$$

Estas ecuaciones nos dan una idea de la complejidad temporal del algoritmo Heap Sort en diferentes casos. En todos los casos, el tiempo de ejecucin es de orden $\Theta(n\log{n})$, lo que hace que Heap Sort sea un algoritmo eficiente para ordenar arreglos grandes.

```
# Average case on random uniform permtations with repetitions of  $n$   elements  $t_a(n) = \frac{3}{4}n^2 + \frac{7}{2}n -\frac{13}{4}= \Theta(n^2) $
import numpy as np
import pandas as pd
import scipy.linalg as la

n = np.array([3.0,4.0,5.0,6.0])    # Number n of sorted elments
t = np.array([14.0,22.75,33.0,44.75]) # Average comparisons + swaps for the corresponding n

data = pd.DataFrame({'n': n, 't': t})
data
```
